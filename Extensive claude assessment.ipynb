Question 48

Write a complete **Student Grade Management System** that:
- Keeps running until user types `quit`
- Asks for student name and score
- Assigns grade
- Saves each entry to `grades.txt`
- Handles all errors

```python
with open("grades.txt", 'w') as gg:
    gg.write("==========Student Grade Management System==========\n")
    while True:
        try:
            s_name = input("Enter the student name: (type 'quit' to stop the program)")
            if s_name == "quit":
                break
            s_score = int(input("Enter student score: "))
        except ValueError:
            print("Please input the correct Value")
            continue
        except Exception as e:
            print(f"The problem is a {e}, Please work on it")
            continue
        if s_score >= 90:
            s_grade="A"
        elif s_score >= 80:
            s_grade= "B"
        elif s_score >= 70:
            s_grade = "C"
        elif s_score >= 60:
            s_grade ="D"
        else:
            s_grade = "F"
        s_info = f"Student name:{s_name}|Student score:{s_score}|Student Grade:{s_grade}"
        gg.write(f"{s_info}\n")
```

Question 50

Asks user to enter income and expenses using a loop
- Stores everything in a list of dictionaries
- Calculates total income, total expenses and balance
- Tells user if they are in profit or loss
- Saves complete report to `finance_report.txt`
- Handles ALL possible errors

```python
with open("finance_report.txt", 'w') as fr:
    fr.write("========== Personal Finance Tracker ==========\n\n")
    calc_list = []
    i = 0
    e = 0
    while True:
        try:
            inc = int(input("What is your income?(type '000' to stop the program) "))
            if inc == 000:
                break
            exp = int(input("What are your expenses? "))
        except ValueError:
            print("Please input only numbers")
            continue
        except Exception as e:
            print(f"The problem is a {e}, Please work on it")
            continue
        calc_list.append({"Income":inc , "Expenses":exp})
        i+=inc
        e+=exp
        tb=inc+exp
        fr.write(f"Income is {inc}, Expenses are {exp}. The total balance here is {tb}\n")
        tb=i+e
    fr.write(f"The total overall Income  is {i}\n")
    fr.write(f"The total overall Expenses  is {e}\n")
    fr.write(f"The total overall balance  is {tb}\n")
    if i>e:
        fr.write("You're in Profit, Good Job")
    else:
        fr.write("You're in Loss, you need to do better")
```
